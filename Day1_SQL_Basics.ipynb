# Two main types of Databases:

Relational databases: SQL Databases
organizes data into one or more tables

*   Each table has column and rows
*   A unique key identifies each row


Relational databases are by far the most popular type of databases. In relational database we are going to store all the data inside the tables. So organize all the data we want to store into these predefined tables and then we can just insert information into there. Its more like a spreadsheet.


Non-Relational databases:(noSQL/ not just SQL)
organize data is anything but a mtraditional table

key-value stores

Documents(XML< JSON, etc)

Graphs

Flexible Tables

Anything that is not relational is non relational database.
